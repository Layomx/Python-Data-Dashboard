#### Numeric Values

```python
# Sales rages (min, max, mean, median, etc.abs)
print("Minimun sales:", df['Sales'].min())
print("Maximun sales:", df['Sales'].max())
print("Sales mean:", df['Sales'].mean())
print("Sales median:", df['Sales'].median())

# Graphics size
plt.figure(figsize = (12, 5))

# Histograms
plt.subplot(1, 2, 1)
sns.histplot(df['Sales'], bins = 50, kde = True, color = 'skyblue')
plt.title("Sales distribution")
plt.xlabel('Sales count')
plt.ylabel('Frequency')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x = df['Sales'], color = 'lightgreen')
plt.title('Sales boxplot')

plt.tight_layout()
plt.show() 
```

This answers a specific question: Are most sales small or are there a few very large sales that dominate the total?
Most of the sales are small, but there is a few big sales impacting the total, this is skewed distribution or long tail

```python
# Plotting a Lorenz Curve

# Sorting the sales
sales = np.sort(df['Sales'].values)

# Calculating accumulated percentage
accumulated_sales = np.cumsum(sales)
accumulated_sales = accumulated_sales / accumulated_sales[-1]

# X axis (transaction percentage)
x = np.linspace(0.0, 1.0, len(sales))

# Plots
plt.figure(figsize = (8, 6))
plt.plot(x, accumulated_sales, label = 'Lorenz Curve', color = 'blue')
plt.plot([0, 1], [0, 1], label = 'Equality line', color = 'red', linestyle = '--')
plt.xlabel('Accumulated transactions percentage')
plt.ylabel('Accumulated sales percentage')
plt.legend()
plt.grid(True)
plt.show()
```

As additional evidence of sales inequality

Further analysis, most transactions are below 500-1000$

##### Another questions
- How many products does a customer typically purchase in one order?
- Are most orders for one unit, or are there bulk purchases?
- Are there suspicious quantities (e.g., zero or very high quantities)?

```python
df.groupby('Order ID')['Quantity'].sum().describe()
```

```python
# Quantity ranges
print("Minimun sales:", df['Quantity'].min())
print("Maximun sales:", df['Quantity'].max())
print("Sales mean:", df['Quantity'].mean())
print("Sales median:", df['Sales'].median())

# Plots
sns.histplot(df.groupby('Order ID')['Quantity'].sum(), bins = 30, kde = True)
plt.title('Products distribution by order')
plt.xlabel('Total quantity per order')
plt.ylabel('Frequency')
plt.show()
```

Based on the analysis, the average order contains 7.56 products, with a standard deviation of 6.22, indicating significant variability. Some bulk purchases skew the distribution, and while no orders contain zero products, there are orders with more than 14 items.

##### Discount questions
- What is the range of discounts applied (0% to what)?
- Are most sales discounted or not?
- Are there extreme discounts that don't make sense (e.g., 90%, 100%)?

```python
# Range of discounts applied in different sales or ordes
print(f"Minimum discount applied of: {df['Discount'].min() * 100:.0f}%")
print(f"Maximum discount applied of: {df['Discount'].max() * 100:.0f}%")
```

```python
df['Discount'].value_counts(normalize = True).sort_values(ascending=False).apply(lambda x: f"{x*100:.2f}%")
```

The range of discounts applied spans from 0% to 80%. While 48.81% of sales occur without any discount, the majority of transactions do include some form of discount commonly around 20%. These discounts are generally moderate, and there are no extreme cases such as 90% or 100% discounts observed in the dataset.
